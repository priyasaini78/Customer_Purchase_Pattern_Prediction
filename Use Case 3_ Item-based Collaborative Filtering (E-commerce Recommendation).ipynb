```python
# Use Case 3: Item-based Collaborative Filtering (E-commerce Recommendation)

# ---------------------------------------------------------------------
# 📦 Import Required Libraries
# ---------------------------------------------------------------------
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
import os

# ---------------------------------------------------------------------


# Read the Excel file
df = pd.read_excel("/Volumes/capston_project-priya_saini(24mbmb34)/default/capston_project-priya_saini(24mbmb34)/ecommerce_data.xlsx", parse_dates=["purchase_date"])

# ---------------------------------------------------------------------
# 🧮 Data Preparation  Compute Implied Ratings
# ---------------------------------------------------------------------
# If 'rating' missing, use 4 for completed purchases and 0 otherwise
df["implied_rating"] = df["rating"].fillna(df["completed"].apply(lambda x: 4 if x == 1 else 0)).astype(float)

# Create User-Item Matrix
user_item = df.pivot_table(
    index="customer_id",
    columns="product_id",
    values="implied_rating",
    aggfunc="mean",
    fill_value=0
)
display(user_item.head())

# ---------------------------------------------------------------------
# 🔍 Compute Item-Item Similarity Matrix
# ---------------------------------------------------------------------
item_sim = cosine_similarity(user_item.T)
item_sim_df = pd.DataFrame(item_sim, index=user_item.columns, columns=user_item.columns)
display(item_sim_df.head())

# ---------------------------------------------------------------------
# 🎯 Define Recommendation Function
# ---------------------------------------------------------------------
def recommend_item_based(user_id, top_n=5):
    """Recommend top N items for a user based on item-item collaborative filtering."""
    if user_id not in user_item.index:
        return []
    user_ratings = user_item.loc[user_id]
    # Weighted score = similarity * user rating
    scores = item_sim_df.dot(user_ratings)
    # Remove already purchased items
    seen = user_ratings[user_ratings > 0].index
    scores = scores.drop(index=seen)
    # Get top N recommendations
    recs = scores.sort_values(ascending=False).head(top_n)
    return recs.index.tolist()

# Example: Top 3 recommendations for 5 users
sample_users = user_item.index.to_list()[:5]
results = {u: recommend_item_based(u, top_n=3) for u in sample_users}
display(results)

# ---------------------------------------------------------------------
# 📊 Visualization Section
# ---------------------------------------------------------------------


# 1 Top 10 Most Purchased Products
top_products = df["product_id"].value_counts().head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_products.index, y=top_products.values, palette="viridis")
plt.title("Top 10 Most Purchased Products")
plt.xlabel("Product ID")
plt.ylabel("Number of Purchases")
plt.tight_layout()
plt.show()

# 2 Customer-Product Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(user_item.iloc[:20, :20], cmap="YlGnBu", cbar=True)
plt.title("Customer vs Product Rating Matrix (Sample 20x20)")
plt.xlabel("Product ID")
plt.ylabel("Customer ID")
plt.tight_layout()
plt.show()

# 3 Item-Item Similarity Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(item_sim_df.iloc[:20, :20], cmap="coolwarm")
plt.title("Item-Item Similarity Heatmap (Sample 20x20)")
plt.xlabel("Product ID")
plt.ylabel("Product ID")
plt.tight_layout()
plt.show()


# ---------------------------------------------------------------------
#  Summary
# ---------------------------------------------------------------------
print("""
 Created an item-based collaborative filtering system using cosine similarity.
 Generated top-N recommendations for sample users.
 Visualized popular products, user-product interactions, and similarity patterns.
 All outputs saved in /FileStore/output/.
""")
```
